```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

import datetime as dt
from matplotlib import dates
import time
```

```python
react = pd.read_csv('dataset/collaborators/npm/react.csv')
angular = pd.read_csv('dataset/collaborators/npm/angular.csv')
print(react.describe())
print(angular.describe())
```

1. Correlation Analysis: correlations between the number of commits, additions, and deletions. A Very strong positive correlation between the number of commits, additions, and deletions. This suggests that collaborators who make more commits also tend to make more additions and deletions.

```python
print(react.corr())
print(angular.corr())
```

2. Top Collaborators

```python
print(react.nlargest(3, 'commits'))
print(angular.nlargest(3, 'commits'))
```

3. Commit Trends Over Time

```python
react.plot(x='name', y='commits', kind='line')
angular.plot(x='name', y='commits', kind='line')
plt.show()
```

4. Additions vs. Deletions: suggesting that collaborators who make more additions also tend to make more deletions.

```python
react.plot(x='additions', y='deletions', kind='scatter')
angular.plot(x='additions', y='deletions', kind='scatter')
plt.show()
```

5. Commit Distribution: We can see that the majority of collaborators have between 1 and 180 commits (React).

```python
react.hist(column='commits', bins=10)
angular.hist(column='commits', bins=10)
plt.show()
```

6. Commit Time distribution: We can see that the distribution of commits over time is relatively symmetrical for each collaborator, suggesting that they made a consistent number of commits throughout the project.

```python
react.plot(kind='density', x='commits')
angular.plot(kind='density', x='commits')
plt.show()
```
